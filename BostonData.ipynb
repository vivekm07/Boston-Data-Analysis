# Deriving a linear regression between MV as output variable and other variables . Print mse and r square. Do with different test sizes:

```python
data.head(9)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
df=data
```

# test size 0.1:

```python
training_features = [ 'CRIM','RM','AGE']
target = 'MV'
X_train, X_test, Y_train, Y_test = train_test_split(df[training_features],
                                                    df[target],
                                                     test_size=0.1,  
                                                   random_state=175)  
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()

model.fit(X_train,Y_train)

predicted=model.predict(X_test)
print("Mean squared error: %.2f"
              % mean_squared_error(Y_test, predicted))
print('R Square score: %.2f' % r2_score(Y_test, predicted))
```

# test size 0.2:

```python
training_features = [ 'CRIM','RM','AGE']
target = 'MV'
X_train, X_test, Y_train, Y_test = train_test_split(df[training_features],
                                                    df[target],
                                                     test_size=0.2,  
                                                   random_state=85) 
```
